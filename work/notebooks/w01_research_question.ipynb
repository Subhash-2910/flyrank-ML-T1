## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

## Research Question

Which pages should be reviewed first for refresh, expansion, protection, pruning, or monitoring based on their historical search performance?

## Decision

The model helps content and SEO teams decide which pages should receive attention first.

## Unit of Analysis

One content page.

## Output

For each page, the system will produce:

- A ranked action queue.
- A suggested action (Refresh, Expand, Protect, Prune, or Monitor).
- Reason codes explaining why the recommendation was made.
- A confidence label indicating how reliable the recommendation is.
- A Model Card summarizing the model, intended use, assumptions, and limitations.

## Action

Content teams can prioritize their limited resources by working on the highest-value pages first instead of reviewing every page manually.

## Cost of a Wrong Recommendation

A false positive recommends work on a page that would not benefit from changes, wasting time and resources.

A false negative fails to recommend a page that could have significantly benefited from intervention, leading to missed traffic and business opportunities.

Therefore, the model should support human decision-making rather than replace it.

## Why Machine Learning?

Page performance depends on many interacting factors such as impressions, clicks, rankings, CTR, search demand, and historical trends. These relationships are difficult to capture using simple rules. Machine learning can learn these patterns from historical data and provide more effective prioritization than manual inspection alone.

## 4. Careful words: what I can and can't claim

*Write what your work will be able to say (observed, directional, decision-support) — and what it never will (causal proof, 'predicting Google').*


This project is designed to **prioritize pages for review**, not to guarantee that a specific action will improve performance. The recommendations should be used as decision support for content and SEO teams rather than as automatic instructions.

A page that is declining in traffic is **not automatically** a good candidate for a refresh. The decline could be caused by seasonality, changes in user intent, increased competition, or search engine algorithm updates. Similarly, a page with stable performance may still deserve protection if it consistently performs well.

To avoid data leakage, the model will only use information that would have been available **before** the prediction period. Metrics from future time windows will never be used as input features.

The model identifies patterns in historical data and estimates which pages are likely to deserve attention. It does **not** establish cause-and-effect relationships, and its recommendations should always be reviewed alongside domain expertise before any content changes are made.
